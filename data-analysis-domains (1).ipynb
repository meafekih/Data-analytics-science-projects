<p style="display: inline-block; padding: 7px; background-color: #ADFF2F; color: #228B22; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: center; border: 1px solid #228B22; font-family: 'Arial', sans-serif;">- Column Descriptions -</p>

- **Date**: The date of the customer transaction.    
- **Customer ID**: Unique identifier for each customer.    
- **Name**: Customer's name.    
- **Email**: Customer's email address.    
- **Age**: Customer's age.    
- **Gender**: Customer's gender.    
- **Transaction Amount**: Amount spent in a single transaction.    
- **Product Category**: Category of the purchased product.    
- **Number of Purchases**: Total number of transactions made by the customer.    
- **Customer Satisfaction**: Rating given by the customer for their satisfaction (on a scale of 1 to 5).
    
<p style="display: inline-block; padding: 7px; background-color: #ADFF2F; color: #228B22; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: center; border: 1px solid #228B22; font-family: 'Arial', sans-serif;">- Principal Studies, Measures, and Information -</p>
    
#### **1. Customer Lifetime Value (CLV):** Customer Lifetime Value predicts the total revenue a business can expect from a customer throughout their entire relationship.    
<p style="display: inline-block; padding: 7px;background-color: #ADFF2F; color: #228B22; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">    
\(\text{CLV} = \text{Transaction Amount} \times \text{Number of Purchases}\)   
</p> 

#### **2. Customer Retention Rate:** Customer Retention Rate calculates the percentage of customers retained over a specific period.
<p style="display: inline-block; padding: 7px; background-color: #ADFF2F; color: #228B22; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
$\text{Customer Retention Rate} = \left(\frac{\text{Number of Customers at End of Period} - \text{Number of New Customers}}{\text{Number of Customers at Start of Period}}\right) \times 100$
</p>   

#### **3. Average Transaction Value (ATV):** Average Transaction Value calculates the average amount spent by a customer in a single transaction.
<p style="display: inline-block; padding: 7px; background-color: #ADFF2F; color: #228B22; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
$\text{ATV} = \frac{\text{Total Revenue}}{\text{Number of Transactions}}$
</p>

#### **4. Customer Satisfaction Score (CSAT):** Customer Satisfaction Score measures the level of satisfaction based on customer ratings.
<p style="display: inline-block; padding: 7px; background-color: #ADFF2F; color: #228B22; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
$\text{CSAT} = \frac{\text{Sum of Customer Satisfaction Ratings}}{\text{Number of Ratings}}$
</p>

#### **5. Customer Churn Rate:** Customer Churn Rate calculates the percentage of customers who stop purchasing or using a service.
<p style="display: inline-block; padding: 7px; background-color: #ADFF2F; color: #228B22;; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
$\text{Churn Rate} = \left(\frac{\text{Number of Customers Lost During a Period}}{\text{Number of Customers at the Start of the Period}}\right) \times 100$    
</p>

#### **6. Purchase Frequency:** Purchase Frequency measures how often customers make a purchase.
<p style="display: inline-block; padding: 7px; background-color: #ADFF2F; color: #228B22; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
$\text{Purchase Frequency} = \frac{\text{Number of Transactions}}{\text{Number of Customers}}$   
</p>

<p style="display: inline-block; padding: 7px; background-color: #fef0e9; color: #AF4343; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: center; border: 1px solid #AF4343; font-family: 'New Times Roman', serif;">- Meta Data -</p>
    
- **Date**: The date on which the advertising data was recorded.
- **Product**: The specific product being advertised. In this example, it could be categorized as 'Product A,' 'Product B,' or 'Product C.'
- **Channel**: The advertising channel through which the campaign was conducted. It could be 'Online,' 'TV,' or 'Radio.'
- **Impressions**: The number of times the ad was viewed or displayed to potential customers.
- **Clicks**: The number of clicks on the ad, indicating user engagement.
- **Conversions**: The number of users who took a desired action, such as making a purchase or filling out a form, in response to the ad.
- **Revenue**: The total revenue generated from the advertising campaign.
- **Cost**: The total cost incurred for running the advertising campaign.
    
    
    
<p style="display: inline-block; padding: 7px; background-color: #fef0e9; color: #AF4343; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: center; border: 1px solid #AF4343; font-family: 'New Times Roman', serif;">-  Metrics -</p>    

#### **1. Return on Investment (ROI)**: Return on Investment measures the profitability of the advertising campaign by evaluating the return relative to the cost incurred.
<p style="display: inline-block; padding: 7px; background-color: #d9ead3; color: #565d54; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
    $\text{ROI} = \frac{(\text{Revenue} - \text{Cost})}{\text{Cost}} \times 100$
    </p>  

    
    
#### **2. Click-Through Rate (CTR)**: Click-Through Rate represents the percentage of people who clicked on the ad after seeing it, providing insights into ad engagement.
<p style="display: inline-block; padding: 7px; background-color: #d9ead3; color: #565d54; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">  
        $\text{CTR} = \frac{\text{Clicks}}{\text{Impressions}} \times 100$    
    </p>      
    
#### **3. Conversion Rate**: Conversion Rate measures the percentage of users who took the desired action after clicking on the ad.

<p style="display: inline-block; padding: 7px; background-color: #d9ead3; color: #565d54; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
     $\text{Conversion Rate} = \frac{\text{Conversions}}{\text{Clicks}} \times 100$
      </p>

#### **4. Cost per Click (CPC)**: Cost per Click calculates the average cost incurred for each click on the ad.

<p style="display: inline-block; padding: 7px; background-color: #d9ead3; color: #565d54; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
     $\text{CPC} = \frac{\text{Cost}}{\text{Clicks}}$
</p>

#### **5. Cost per Conversion**: Cost per Conversion assesses the average cost required to achieve a conversion.

<p style="display: inline-block; padding: 7px; background-color: #d9ead3; color: #565d54; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
     $\text{Cost per Conversion} = \frac{\text{Cost}}{\text{Conversions}}$
      </p>

#### **6. Revenue per Click**: Revenue per Click evaluates the average revenue generated for each click on the ad.

<p style="display: inline-block; padding: 7px; background-color: #d9ead3; color: #565d54; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
     $\text{Revenue per Click} = \frac{\text{Revenue}}{\text{Clicks}}$
      </p>

- **Date**: The date of the HR record.
- **Employee ID**: Unique identifier for each employee.
- **Name**: Employee's name.
- **Department**: Department in which the employee works.
- **Position**: Employee's job position.
- **Salary**: Employee's salary.
- **Years of Experience**: Number of years of work experience.
- **Performance Rating**: Employee's performance rating (on a scale of 1 to 5).
- **Training Hours**: Number of hours of training received.
- **Employee Satisfaction**: Rating given by the employee for their satisfaction (on a scale of 1 to 5).

<p style="display: inline-block; padding: 7px; background-color: #D8BFD8; color: #800080; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: center; border: 1px solid #800080; font-family: 'Arial', sans-serif;">- Principal Studies, Measures, and Information -</p>

#### **1. Employee Turnover Rate:** Employee Turnover Rate calculates the percentage of employees who leave the company within a specified period.
<p style="display: inline-block; padding: 7px; background-color: #D8BFD8; color: #800080; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
$\text{Turnover Rate} = \left(\frac{\text{Number of Employees Left}}{\text{Average Number of Employees}}\right) \times 100$ 
</p>     

#### **2. Employee Productivity:** Employee Productivity measures the efficiency and output of employees.
<p style="display: inline-block; padding: 7px; background-color: #D8BFD8; color: #800080; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
$\text{Productivity} = \frac{\text{Output}}{\text{Input}}$
</p>  

#### **3. Compensation Ratio:** Compensation Ratio evaluates the relationship between an employee's salary and the midpoint of the salary range for their position.
<p style="display: inline-block; padding: 7px; background-color: #D8BFD8; color: #800080; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
$\text{Compensation Ratio} = \frac{\text{Employee Salary}}{\text{Midpoint Salary}} \times 100$
</p>  

#### **4. Training Effectiveness:** Training Effectiveness assesses the impact of training programs on employee performance.
<p style="display: inline-block; padding: 7px; background-color: #D8BFD8; color: #800080; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
$\text{Training Effectiveness} = \frac{\text{Improvement in Performance}}{\text{Cost of Training}}$    
</p>      

#### **5. Employee Satisfaction Index:** Employee Satisfaction Index aggregates satisfaction ratings to measure overall employee satisfaction.
<p style="display: inline-block; padding: 7px; background-color: #D8BFD8; color: #800080; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
$\text{Employee Satisfaction Index} = \frac{\text{Sum of Satisfaction Ratings}}{\text{Number of Employees}}$    
</p>      

#### **6. Promotion Rate:** Promotion Rate calculates the percentage of employees who were promoted within a specific period.
<p style="display: inline-block; padding: 7px; background-color: #D8BFD8; color: #800080; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
$\text{Promotion Rate} = \left(\frac{\text{Number of Promotions}}{\text{Average Number of Employees}}\right) \times 100$
</p>

- **Date**: The date on which the marketing data was recorded.
- **Campaign**: The specific marketing campaign.
- **Platform**: The advertising platform used for the campaign, such as Facebook, Instagram, or Google Ads.
- **Impressions**: The number of times the ad was viewed.
- **Clicks**: The number of clicks on the ad, indicating user engagement.
- **Conversions**: The number of users who took a desired action in response to the ad.
- **Revenue**: The total revenue generated from the marketing campaign.
- **Cost**: The total cost incurred for running the marketing campaign.

<p style="display: inline-block; padding: 7px; background-color: #E6E6FA; color: #4682B4; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: center; border: 1px solid #4682B4; font-family: 'Arial', sans-serif;">- Principal Studies, Measures, and Information -</p>

#### **1. Return on Investment (ROI):** Return on Investment measures the profitability of the marketing campaign by evaluating the return relative to the cost incurred.
<p style="display: inline-block; padding: 7px; background-color: #d9ead3; color: #565d54; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
$\text{ROI} = \frac{(\text{Revenue} - \text{Cost})}{\text{Cost}} \times 100$
</p>  

#### **2. Click-Through Rate (CTR):** Click-Through Rate represents the percentage of people who clicked on the ad after seeing it.
<p style="display: inline-block; padding: 7px; background-color: #d9ead3; color: #565d54; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
$\text{CTR} = \frac{\text{Clicks}}{\text{Impressions}} \times 100$
</p>

#### **3. Conversion Rate:**
Conversion Rate measures the percentage of users who took the desired action after clicking on the ad.
<p style="display: inline-block; padding: 7px; background-color: #d9ead3; color: #565d54; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
$\text{Conversion Rate} = \frac{\text{Conversions}}{\text{Clicks}} \times 100$
</p>    

#### **4. Cost per Click (CPC):** : Cost per Click calculates the average cost incurred for each click on the ad.
<p style="display: inline-block; padding: 7px; background-color: #d9ead3; color: #565d54; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
$\text{CPC} = \frac{\text{Cost}}{\text{Clicks}}$
</p>

#### **5. Cost per Conversion:** Cost per Conversion assesses the average cost required to achieve a conversion.
<p style="display: inline-block; padding: 7px; background-color: #d9ead3; color: #565d54; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
$\text{Cost per Conversion} = \frac{\text{Cost}}{\text{Conversions}}$
</p>

#### **6. Revenue per Click:** Revenue per Click evaluates the average revenue generated for each click on the ad.
<p style="display: inline-block; padding: 7px; background-color: #d9ead3; color: #565d54; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
$\text{Revenue per Click} = \frac{\text{Revenue}}{\text{Clicks}}$
</p>

- **Date**: The date on which the financial data was recorded.
- **Company**: The name of the company.
- **Revenue**: The total income generated by the company.
- **Expenses**: The total costs incurred by the company.
- **Profit**: The net profit (Revenue - Expenses).
- **Assets**: The total value of assets owned by the company.
- **Liabilities**: The total amount of debts and obligations.
- **Net Income**: The company's total income after deducting all expenses.
- **EPS (Earnings Per Share)**: The earnings generated per outstanding share of stock.

<p style="display: inline-block; padding: 7px; background-color: #B0E0E6; color: #008080; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: center; border: 1px solid #008080; font-family: 'Arial', sans-serif;">- Principal Studies, Measures, and Information -</p>

#### **1. Profit Margin:** Profit Margin measures the percentage of profit generated relative to revenue.
<p style="display: inline-block; padding: 7px; background-color: #d9ead3; color: #565d54; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
$\text{Profit Margin} = \frac{\text{Profit}}{\text{Revenue}} \times 100$
</p>
    
#### **2. Return on Assets (ROA):** Return on Assets evaluates how efficiently a company uses its assets to generate profit.
<p style="display: inline-block; padding: 7px; background-color: #d9ead3; color: #565d54; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
$\text{ROA} = \frac{\text{Net Income}}{\text{Assets}} \times 100$
</p>

#### **3. Return on Equity (ROE):** Return on Equity measures the profitability of a company in relation to shareholders' equity.
<p style="display: inline-block; padding: 7px; background-color: #d9ead3; color: #565d54; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
$\text{ROE} = \frac{\text{Net Income}}{\text{Shareholders' Equity}} \times 100$
</p>

#### **4. Earnings Per Share (EPS):** Earnings Per Share indicates the portion of a company's profit allocated to each outstanding share of common stock.
<p style="display: inline-block; padding: 7px; background-color: #d9ead3; color: #565d54; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
$\text{EPS} = \frac{\text{Net Income}}{\text{Number of Outstanding Shares}}$
</p>

#### **5. Debt-to-Equity Ratio:** Debt-to-Equity Ratio assesses the proportion of debt used to finance a company's assets relative to shareholders' equity. 
<p style="display: inline-block; padding: 7px; background-color: #d9ead3; color: #565d54; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
$\text{Debt-to-Equity Ratio} = \frac{\text{Liabilities}}{\text{Shareholders' Equity}}$
</p>  

#### **6. Operating Income Margin:** Operating Income Margin measures the percentage of operating income relative to revenue.
<p style="display: inline-block; padding: 7px; background-color: #d9ead3; color: #565d54; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
$\text{Operating Income Margin} = \frac{\text{Operating Income}}{\text{Revenue}} \times 100$  
</p>

- **Date**: The date of the social media post.
- **Post ID**: Unique identifier for each social media post.
- **Author**: Author or creator of the post.
- **Content**: The content or message of the post.

- **Comments**: Number of comments received.

- **Sentiment**: Sentiment of the post (Positive, Neutral, Negative).

<p style="display: inline-block; padding: 7px; background-color: #00CED1; color: #008B8B; text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: center; border: 1px solid #008B8B; font-family: 'Arial', sans-serif;">- Principal Studies, Measures, and Information -</p>

#### **1. Engagement Rate:** Engagement Rate measures the level of interaction with the content.
<p style="display: inline-block; padding: 7px; background-color: #00CED1; color: #008080;  text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
$\text{Engagement Rate} = \frac{\text{Likes} + \text{Shares} + \text{Comments}}{\text{Total Followers}} \times 100$
</p>   

#### **2. Hashtag Popularity:** Hashtag Popularity assesses the reach and popularity of hashtags.
<p style="display: inline-block; padding: 7px; background-color: #00CED1; color: #008080;  text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
$\text{Hashtag Popularity} = \frac{\text{Number of Posts with Hashtag}}{\text{Total Number of Posts}} \times 100$ 
</p>   

#### **3. Sentiment Analysis:** Sentiment Analysis categorizes posts into positive, neutral, or negative sentiments.

#### **4. Virality Index:** Virality Index measures the likelihood of content being shared.
<p style="display: inline-block; padding: 7px; background-color: #00CED1; color: #008080;  text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
$\text{Virality Index} = \frac{\text{Shares}}{\text{Total Views}} \times 100$ 
</p>      
 
#### **5. Click-Through Rate (CTR):** Click-Through Rate represents the percentage of people who clicked on the post.
<p style="display: inline-block; padding: 7px; background-color: #00CED1; color: #008080;  text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
$\text{CTR} = \frac{\text{Number of Clicks}}{\text{Number of Impressions}} \times 100$   
</p>      

#### **6. Follower Growth Rate:** Follower Growth Rate calculates the percentage increase in followers.
<p style="display: inline-block; padding: 7px; background-color: #00CED1; color: #008080;  text-decoration: none; border-radius: 5px 5px; font-size: 150%; text-align: left; border: 1.5px solid #828c7e; font-family: 'New Times Roman', serif;">
$\text{Follower Growth Rate} = \frac{\text{New Followers}}{\text{Total Followers at the Beginning}} \times 100$  
</p>
